```python
# Importing required Libraries in start

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
# Reading cars.csv file using pandas

cr=pd.read_csv('/content/drive/MyDrive/Datasets/cars (1).csv')
cr
```

```python
# To check shape of our dataset

cr.shape
```

```python
# To check info of our dataset

cr.info()
```

```python
# To check null values in our dataset

cr.isnull().sum()
```

```python
# To check duplicate values in our dataset

cr.duplicated().value_counts()
```

```python
# The column having contains some imporper value : i.e "?"

cr['normalized-losses'].value_counts()
```

```python
# Replacing "?" value into NaN value

cr['normalized-losses'].replace('?',np.nan,inplace=True)
```

```python
# Updating column data type into float and replacing NaN value into column's mean value

cr['normalized-losses']=cr['normalized-losses'].astype('float')
cr['normalized-losses'].fillna(cr['normalized-losses'].mean(),inplace=True)
```

```python
# After doing some Experiments our column is proper now

cr['normalized-losses'].value_counts()
```

```python
# Same improper element having in this column

cr['horsepower'].value_counts()
```

```python
# Doing same steps on ["horsepower"] column as we done for previous column


cr['horsepower'].replace('?',np.nan,inplace=True)
cr['horsepower']=cr['horsepower'].astype('float')
cr['horsepower'].fillna(cr['horsepower'].mean(),inplace=True)
```

```python
# Our column is proper now

cr['horsepower'].value_counts()
```

```python
# Checking for outliers using seaborn boxplot

plt.figure(figsize=(9,5))
sn.boxplot(data=cr,x='price',y='make')
plt.xlabel('Car Maker')
plt.ylabel('Car Price Range')
plt.title('Car price Outliers')
plt.show()
```

In boxplot we will see there are some outliers, b'coz they are very few so we are not removing them otherwise it will affect on our model. As considering they are valid data points

```python
# Separating numerical columns from cars data set and stored in variable num_col

num_col=cr.select_dtypes(exclude='object')
num_col
```

```python
# Separating categorical columns from cars data set and stored in variable cat_col

cat_col=cr.select_dtypes(include='object')
cat_col
```

```python
 # Applying Label Encoder on cat_col

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in cat_col:
  le = LabelEncoder()
  cat_col[col] = le.fit_transform(cat_col[col])
```

```python
# Now cat_col can be calculated

cat_col
```

```python
num_col
```

```python
# Again num_col & cat_col is concatinated

cr_cln = pd.concat([num_col,cat_col],axis=1)
cr_cln
```

```python
# Taken x as our feature for model

x=cr_cln.drop(columns='price')
x
```

```python
x.isnull().sum()
```

```python
# Taken y as target for our model

y=cr_cln['price']
y
```

```python
# plotting scatterplot between x columns with price column

for col in x:
  print(col)
  plt.scatter(x[col],y)
  plt.title(f'Scatterplot between {col} and price')
  plt.show()
```

```python
# To check which feature is more important in x
from sklearn.ensemble import ExtraTreesRegressor
etr = ExtraTreesRegressor()
etr.fit(x,y)
```

```python
print(etr.feature_importances_)
```

```python
#plotting graph of features in max importance order

fe = pd.Series(etr.feature_importances_,index=x.columns)
fe.nlargest(14).plot(kind='barh')
plt.show()
```

```python
# Spliting data into train & test for model

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=17)
```

```python
# Used MinMaxscaler to minimize the scaling of our data points

from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

xtrain = mm.fit_transform(xtrain)
xtest = mm.fit_transform(xtest)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
# Applying Linear Regression on data
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(xtrain,ytrain)

pred = lr.predict(xtest)

r2_score(ytest,pred)*100
```

**By using Linear Regression model it is giving 70.5% accuracy, which is low**

```python
# Using KFold Cross valadation techinque

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

lr = LinearRegression()
kfold = KFold(10)
result = cross_val_score(lr,xtrain,ytrain,cv=kfold)*100
print(result)
print(np.mean(result))
```

**After applying kfold technique it is giving average accuracy of 73.2%, which is still low**

# **Ridge**

```python
# Applying Ridge for model using Hyperparameters

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
R =  Ridge()
param_grid = {'alpha':[0.001,0.1,1,10,15,20,25,30,50,100]}

r_model = GridSearchCV(R,param_grid,scoring='neg_mean_squared_error',cv=5)
r_model.fit(xtrain,ytrain)
```

```python
print(r_model.best_params_)
```

# **Lasso Model**

```python
# Applying Lasso for model using Hyperparameters

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
L =  Lasso()
param_grid = {'alpha':[0.001,0.1,1,10,15,20,25,30,50,100]}

l_model = GridSearchCV(L,param_grid,scoring='neg_mean_squared_error',cv=5)
l_model.fit(xtrain,ytrain)
```

```python
print(l_model.best_params_)
```

```python
ridge_pred = r_model.predict(xtest)
lasso_pred = l_model.predict(xtest)
```

# **Final Testing with Accurracy**

```python
# Using final best parameter to check accuracy

from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)
ridge.fit(xtrain,ytrain)

ridge.score(xtest,ytest)*100
```

```python
# Using final best parameter to check accuracy

from sklearn.linear_model import Lasso

lasso = Lasso(alpha=25)
lasso.fit(xtrain,ytrain)

lasso.score(xtest,ytest)*100
```
